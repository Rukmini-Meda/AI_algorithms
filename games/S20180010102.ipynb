### Class to define a generic game

```python
class Game(object):
    
#     Constructor of the game class
    def __init__(self,size,con_size,depth):
        self.board=[]
        for i in range(size):
            self.board.append(['_']*size)
        self.size=size
        self.con_size=con_size
        self.player='x'
        self.computer='o'
        self.depthcutoff=depth
        
#   Cutoff function used in Depth limited minimax
    def cutoff(self,depth):
#         It returns heuristic value after a particular depth
        if depth>=self.depthcutoff:
            return True
        else:
            return False

#   Utility function that will be overrided by specific game classes 
    def utility(self):
        pass
    
#   Heuristic evaluation function that will be overrided by specific game classes
    def heuristic_evaluation_function(self):
        pass
    
#     Creative heuristic for required path that returns value based on depth and heuristic
    def creative_evaluation_function(self,depth):
        return depth+self.heuristic_evaluation_function()

#    Function to find whether any moves are left
    def movesLeft(self):
        for i in range(self.size):
            for j in range(self.size):
                if self.board[i][j]=='_':
                    return True
        return False
    
#   Function to implement Minimax algorithm
    def minimax1(self,isMax):
        
#         Terminal conditions
        score=self.utility()
        if score==1 or score==-1:
            return score
        else:
            if self.movesLeft()==False:
                return 0
            else:
#                 Recursively finding out the search tree based on turn of the player
#                 Computer's turn
                if isMax:
                    best=-1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.computer
                                temp=self.minimax1(~isMax)
                                best=max(best,temp)
                                self.board[i][j]='_' 
#                 Player's turn
                else:
                    best=1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.player
                                temp=self.minimax1(~isMax)
                                best=min(best,temp)
                                self.board[i][j]='_'
                return best
            
#     Function to implement Minimax with alpha beta pruning
    def minimax2(self,isMax,alpha,beta):
        
#         Terminal conditions
        score=self.utility()
        if score==1 or score==-1:
            return score
        else:
            if self.movesLeft()==False:
                return 0
            else:
#                 Recursively finding out the search tree based on player's turn
#                 Computer's turn
                if isMax:
                    best=-1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.computer
                                temp=self.minimax2(~isMax,alpha,beta)
                                best=max(best,temp)
                                
#                                 Alpha beta modifications
                                alpha=max(alpha,best)
                                self.board[i][j]='_'
                                if beta<=alpha:
                                    break
#               Player's turn
                else:
                    best=1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.player
                                temp=self.minimax2(~isMax,alpha,beta)
                                best=min(best,temp)
                                
#                                 Alpha beta modifications
                                beta=min(beta,best)
                                self.board[i][j]='_'
                                if beta<=alpha:
                                    break
                return best
    
#     Function to implement Depth Limited Minimax
    def minimax3(self,depth,isMax):
        
#         Cutoff condition to return heuristic value
        if self.cutoff(depth):
            temp=self.heuristic_evaluation_function()
            return temp
        else:
            if self.movesLeft()==False:
                return 0
            else:
#                 Searching the tree recursively based on players turn
#                 Computer's turn
                if isMax:
                    best=-1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.computer
                                temp=self.minimax3(depth+1,~isMax)
                                best=max(best,temp)
                                self.board[i][j]='_'
#               Player's turn
                else:
                    best=1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.player
                                temp=self.minimax3(depth+1,~isMax)
                                best=min(best,temp)
                                self.board[i][j]='_'
                return best
            
#     Function to implement Depth Limited Minimax with alpha beta pruning
    def minimax4(self,depth,isMax,alpha,beta):
        
#         Cutoff condition
        if self.cutoff(depth):
            temp=self.heuristic_evaluation_function()
            return temp
        else:
            if self.movesLeft()==False:
                return 0
            else:
#                 Searching the tree recursively 
#                Computer's turn
                if isMax:
                    best=-1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.computer
                                temp=self.minimax4(depth+1,~isMax,alpha,beta)
                                best=max(best,temp)
#                                 Alpha beta modifications
                                alpha=max(alpha,best)
                                self.board[i][j]='_'
                                if beta<=alpha:
                                    break
                else:
                    best=1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.player
    #                             print(i,j)
    #                             print(best)
                                temp=self.minimax4(depth+1,~isMax,alpha,beta)
    #                             print(best,i,j,temp)
                                best=min(best,temp)
#                                Alpha beta modifications
                                beta=min(beta,best)
                                self.board[i][j]='_'
                                if beta<=alpha:
                                    break
                                
                return best
            
#       Function to implement Depth limited minimax with my very own heuristic function 
#        using alpha beta pruning
    def minimax5(self,depth,isMax,alpha,beta):
        
#         Cutoff condition
        if self.cutoff(depth):
            temp=self.creative_evaluation_function()
            return temp
        else:
            if self.movesLeft()==False:
                return 0
            else:
                if isMax:
                    best=-1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.computer
                                temp=self.minimax4(depth+1,~isMax,alpha,beta)
                                best=max(best,temp)
                                
#                                 Alpha beta modifications
                                alpha=max(alpha,best)
                                self.board[i][j]='_'
                                if beta<=alpha:
                                    break
                                
                else:
                    best=1000
                    for i in range(self.size):
                        for j in range(self.size):
                            if self.board[i][j]=='_':
                                self.board[i][j]=self.player
                                temp=self.minimax4(depth+1,~isMax,alpha,beta)
                                best=min(best,temp)
                                
#                                 Alpha beta modifications
                                beta=min(beta,best)
                                self.board[i][j]='_'
                                if beta<=alpha:
                                    break                         
                return best

# Function to find next best move based onappropriate algorithm
    def FindBestMove(self,choice):
        best_move_val=-1000
        best_move=Move(-1,-1)
        for i in range(self.size):
            for j in range(self.size):
                if self.board[i][j]=='_':
                    self.board[i][j]=self.computer
                    if choice==1:
                        
#                         Minimax
                        move_val=self.minimax1(False)
                    elif choice==2:
            
#                        Minimax with alpha beta pruning
                        move_val=self.minimax2(False,-1000,1000)
                    elif choice==3:
            
#                        Depth limited minimax 
                        move_val=self.minimax3(0,False)
                    elif choice==4:
            
#                          Depth limited minimax with alpha beta pruning
                        move_val=self.minimax4(0,False,-1000,1000)
                    elif choice==5:
            
#                       Depth limited minimax with custom heuristic function and alpha beta pruning       
                        move_val=self.minimax5(0,False,-1000,1000)
                    self.board[i][j]='_'
                    if move_val>best_move_val:
                        best_move_val=move_val
                        best_move.x=i
                        best_move.y=j
        return best_move

#  Utility Function to find next move in a game
    def game_move(self,pos,choice):
        print("================================================================")
        self.player='x'
        self.computer='o'
        x,y=pos[0],pos[1]
        self.board[x][y]=self.player
        self.print_board()
        output=0
        print("================================================================")
        start_time=time.time()
        move_computer=self.FindBestMove(choice)
        end_time=time.time()
        self.board[move_computer.x][move_computer.y]=self.computer
        self.print_board()
        print("Time taken by the computer for the move is ",end_time-start_time,"s")
        print("================================================================")

        return move_computer
    
#  Utility function to clear the board
    def clear_board(self):
        for i in range(self.size):
            self.board=[]
            self.board.append(['_']*self.size)

#             Utility function to print the board
    def print_board(self):
        for i in range(self.size):
            print(self.board[i])
```

### Class to define a TicTacToe game extending the generic game

```python
class TicTacToe(Game):

#     Constructor
    def __init__(self):
        super(TicTacToe,self).__init__(3,3,3)
        
#      Function to implement heuristic evaluation function
    def heuristic_evaluation_function(self):
        self.player='x'
        self.computer='o'
        heuristic=0
        
#         Calculating heuristic for a row for 3 matches
        for i in range(self.size):
            if self.board[i][0]==self.board[i][1] and self.board[i][1]==self.board[i][2]:
                if self.board[i][0]==self.player:
                    heuristic-=1000
                elif self.board[i][0]==self.computer:
                    heuristic+=1000
                    
#                     Calculating heuristic for a column for 3 matches
        for j in range(self.size):
            if self.board[0][j]==self.board[1][j] and self.board[1][j]==self.board[2][j]:
                if self.board[0][j]==self.player:
                    heuristic-=1000
                elif self.board[0][j]==self.computer:
                    heuristic+=1000
                    
#             Calculating heuristic for diagonals for 3 matches
        if self.board[0][0]==self.board[1][1] and self.board[1][1]==self.board[2][2]:
            if self.board[0][0]==self.player:
                heuristic-=1000
            elif self.board[0][0]==self.computer:
                heuristic+=1000
                

        if self.board[0][2]==self.board[1][1] and self.board[1][1]==self.board[2][0]:
            if self.board[0][2]==self.player:
                heuristic-=1000
            elif self.board[0][2]==self.computer:
                heuristic+=1000
                
#         Calculating heuristic for rows for 2 matches
        for k in range(self.size):
            if self.board[k][0]==self.board[k][1] and self.board[k][2]=='_':
                if self.board[k][0]==self.player:
                    heuristic-=10
                elif self.board[k][0]==self.computer:
                    heuristic+=10
            elif self.board[k][1]==self.board[k][2] and self.board[k][0]=='_':
                if self.board[k][1]==self.player:
                    heuristic-=10
                elif self.board[k][1]==self.computer:
                    heuristic+=10
            if self.board[k][0]==self.board[k][1] and self.board[k][2]=='o' and self.board[k][0]=='x':
                heuristic+=100
            elif self.board[k][1]==self.board[k][2] and self.board[k][0]=='o' and self.board[k][1]=='x':
                heuristic+=100
            elif self.board[k][0]==self.board[k][2] and self.board[k][0]=='x' and self.board[k][1]=='x':
                heuristic+=100
                
#         Calculating heuristic for columns for 2 matches
        for l in range(self.size):
            if self.board[0][l]==self.board[1][l] and self.board[2][l]=='_':
                if self.board[0][l]==self.player:
                    heuristic-=10
                elif self.board[0][l]==self.computer:
                    heuristic+=10
            elif self.board[1][l]==self.board[2][l] and self.board[0][l]=='_':
                if self.board[1][l]==self.player:
                    heuristic-=10
                elif self.board[1][l]==self.computer:
                    heuristic+=10
            if self.board[0][l]==self.board[1][l] and self.board[2][l]=='o' and self.board[0][l]=='x':
                heuristic+=100
            elif self.board[1][l]==self.board[2][l] and self.board[1][l]=='x' and self.board[0][l]=='o':
                heuristic+=100
            elif self.board[0][l]==self.board[2][l] and self.board[0][l]=='x' and self.board[1][l]=='o':
                heuristic+=100
                
#       Calculating heuristic for diagonals for 2 matches and blocking condition
        if self.board[0][0]==self.board[1][1] and self.board[0][0]=='x' and self.board[1][1]=='o':
            heuristic+=100
        elif self.board[0][0]==self.board[2][2] and self.board[0][0]=='x' and self.board[1][1]=='o':
            heuristic+=100
        elif self.board[1][1]==self.board[2][2] and self.board[1][1]=='x' and self.board[0][0]=='o':
            heuristic+=100
        if self.board[0][2]==self.board[1][1] and self.board[0][2]=='x' and self.board[2][0]=='o':
            heuristic+=100
        elif self.board[2][0]==self.board[1][1] and self.board[2][0]=='x' and self.board[0][2]=='o':
            heuristic+=100
        elif self.board[0][2]==self.board[2][0] and self.board[0][2]=='x' and self.board[1][1]=='o':
            heuristic+=100
            
#             Calculating heuristic for diagonals for 2 matches
        if (self.board[0][0]==self.board[1][1] and self.board[2][2]=='_') or (self.board[1][1]==self.board[2][2] and self.board[0][0]=='_'):
            if self.board[1][1]==self.player:
                heuristic-=10
            elif self.board[1][1]==self.computer:
                heuristic+=10
        if (self.board[0][2]==self.board[1][1] and self.board[2][0]=='_') or (self.board[1][1]==self.board[2][0] and self.board[0][2]=='_'):
            if self.board[1][1]==self.player:
                heuristic-=10
            elif self.board[1][1]==self.computer:
                heuristic+=10
                
#                 Calculating heuristic for diagonals for 1 match
        if (self.board[0][0]==self.board[1][1] and self.board[0][0]=='_'):
            if self.board[2][2]==self.player:
                heuristic-=1
            elif self.board[2][2]==self.computer:
                heuristic+=1
        elif (self.board[1][1]==self.board[2][2] and self.board[1][1]=='_'):
            if self.board[0][0]==self.player:
                heuristic-=1
            elif self.board[0][0]==self.computer:
                heuristic+=1
        if self.board[0][2]==self.board[1][1] and self.board[1][1]=='_':
            if self.board[2][0]==self.player:
                heuristic-=1
            elif self.board[2][0]==self.computer:
                heuristic+=1
        elif self.board[1][1]==self.board[2][0] and self.board[1][1]=='_':
            if self.board[0][0]==self.player:
                heuristic-=1
            elif self.board[0][0]==self.computer:
                heuristic+=1
            
#             Calculating heuristic for rows for 1 match
        for r in range(self.size):
            count=0
            for c in range(self.size):
                if self.board[r][c]=='_':
                    count+=1
                else:
                    key=self.board[r][c]
            if count==2:
                if key==self.player:
                    heuristic-=1
                elif key==self.computer:
                    heuristic+=1
                    
#       Calculating heuristic for columns for 1 match
        for g in range(self.size):
            count=0
            for h in range(self.size):
                if self.board[h][g]=='_':
                    count+=1
                else:
                    key=self.board[h][g]
            if count==2:
                if key==self.player:
                    heuristic-=1
                elif key==self.computer:
                    heuristic+=1     
        return heuristic
        
#     Utility function of Tic Tac Toe    
    def utility(self):
        for i in range(self.size):
            
#             Row checks
            if self.board[i][0]==self.board[i][1] and self.board[i][1]==self.board[i][2]:
                if self.board[i][0]==self.player:
                    return  -1
                elif self.board[i][0]==self.computer:
                    return 1
                
#                 Column checks
        for j in range(self.size):
            if self.board[0][j]==self.board[1][j] and self.board[1][j]==self.board[2][j]:
                if self.board[0][j]==self.player:
                    return -1
                elif self.board[0][j]==self.computer:
                    return 1
                
#                 Diagonal checks
        if self.board[0][0]==self.board[1][1] and self.board[1][1]==self.board[2][2]:
            if self.board[0][0]==self.player:
                return -1
            elif self.board[0][0]==self.computer:
                return 1
        if self.board[0][2]==self.board[1][1] and self.board[1][1]==self.board[2][0]:
            if self.board[0][2]==self.player:
                return -1
            elif self.board[0][2]==self.computer:
                return 1
            
        return 0
```

### Class to define Open field Tic Tac Toe game

```python
class OpenFieldTicTacToe(Game):
    
#     Constructor
    def __init__(self,size,con_size,depth):
        super(OpenFieldTicTacToe,self).__init__(size,con_size,depth)
        
#         Utility function of open field tic tac toe
    def utility(self):
        flag=0
        
#         Row checks
        for i in range(self.size):
            front=0
            rear=front+self.con_size-1
            while rear<self.size:
                flag=0
                for k in range(front,rear):
                    if self.board[i][k]!=self.board[i][k+1]:
                        flag=1
                        break
                if flag!=1:
                    if self.board[i][front]==self.player:
                        return -1
                    elif self.board[i][front]==self.computer:
                        return 1
                front+=1
                rear+=1
        flag=0
#         Column checks
        for j in range(self.size):
            front=0
            rear=front+self.con_size-1
            while rear<self.size:
                flag=0
                for l in range(front,rear):
                    if self.board[l][j]!=self.board[l+1][j]:
                        flag=1
                        break
                if flag!=1:
                    if self.board[front][j]==self.player:
                        return -1
                    elif self.board[front][j]==self.computer:
                        return 1
                front+=1
                rear+=1

        front=0
#         Diagonal checks
        rear=front+self.con_size-1
        while rear<self.size:
            flag=0
            for l in range(front,rear):
                if self.board[l][l]!=self.board[l+1][l+1]:
                    flag=1
                    break
            if flag!=1:
                if self.board[front][front]==self.player:
                    return -1
                elif self.board[front][front]==self.computer:
                    return 1
            front+=1
            rear+=1

        temp=[self.board[i][size-i-1] for i in range(self.size)]
        front=0
        rear=front+self.con_size-1
        while rear<self.size:
            flag=0
            for l in range(front,rear):
                if temp[l]!=temp[l+1]:
                    flag=1
                    break
            if flag!=1:
                if temp[front]==self.player:
                    return -1
                elif temp[front]==self.computer:
                    return 1
            front+=1
            rear+=1

        return 0
    
#     Utility heuristic function for open field tic tac toe
    def Heuristic(self, arr):
        score=0
        if arr[0]==self.player:
            score=1
        elif arr[0]==self.computer:
            score=-1
            
        for i in range(1,self.size):
            if arr[i]==self.player:
                if score>=1:
                    score*=10
                elif score<=-1:
                    score=0
                else:
                    score=-1
            elif arr[i]==self.computer:
                if score<=-1:
                    score*=10
                elif score>=1:
                    return 0
                else:
                    score=-1
        
        return score
            
#    Heuristic evaluation function for open field tic tac toe
    def heuristic_evaluation_function(self):
        
        score=0
        temp=self.board
        
        for i in range(self.size):
            score+=self.Heuristic(temp[i])
        
        for i in range(self.size):
            col=[]
            for j in range(self.size):
                col.append(temp[i][j])
            score+=self.Heuristic(col)
            
        diag1=[]
        for i in range(self.size):
            diag1.append(temp[i][j])
        score+=self.Heuristic(diag1)
        
        diag2=[]
        for i in range(self.size):
            for j in range(self.size):
                if i+j==self.size-1:
                    diag2.append(temp[i][j])
        score+=self.Heuristic(diag2)
        
        return score
        
                
        
    
```

### Demonstrating Open field tic tac toe

##### (i) Demonstrating Minimax

```python
print("Enter the board size:")
size=int(input())
print("Enter the number of connecting pieces to win:")
con_size=int(input())
ticgame=OpenFieldTicTacToe(size,con_size,4)
print("Which algo would you like to play against?")
print("1 for Minimax")
print("2 for Minimax with alpha beta pruning")
print("3 for Depth limited minimax")
print("4 for Depth limited minimax with alpha beta pruning")
print("5 for Depth limited minimax with custom heuristic")
choice=int(input())
print("================================================================")
ticgame.print_board()
print("================================================================")
playing=True
while playing:
    x,y=map(int,input().split())
    ticgame.board[x][y]='x'
    if ticgame.utility()==1 or ticgame.utility()==-1 or (ticgame.utility()==0 and ticgame.movesLeft()==False):
        playing=False
        output=ticgame.utility()
        break
    ticgame.board[x][y]='_'
    ticgame.game_move([x,y],choice)
    if ticgame.utility()==1 or ticgame.utility()==-1 or (ticgame.utility()==0 and ticgame.movesLeft()==False):
        playing=False
        output=ticgame.utility()
        break
if output==1:
    print("You've lost. Better luck next time!!")
elif output==-1:
    print("Congratulations, you've won!!")
else:
    print("Its a DRAW!!")
```
